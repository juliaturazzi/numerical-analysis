Função:

```python
# Ta = 25°C; k = -1,335.
def f(x, y): 
  return (-1.355 * (y -25))
```

Obs: A função que contém a solução analítica se encontra antes dos outros métodos por ser utilizada para adquirir os erros relativos de cada um deles, solicitados na questão (d).
---



Solução Analítica:
---


```python
x0 = 0
x = 15

h = 0.25

n = round((x - x0) / h)
```

```python
def analitica(x):
  return (25 + (6 -25) * (math.e)**(-1.335 * x))
```

```python
p1_a = []
p2_a = []

for i in range(1, 60):
  p1_a.append(i * h)
  p2_a.append(analitica(i * h))
```

```python
plt.plot(p1_a, p2_a, 'r-', label = "Analítica")

plt.xlabel('Tempo (min)')
plt.ylabel('Temperatura (°C)')
plt.title('Analítica')
plt.grid(False)

plt.show()
```

Método de Euler:
---



```python
def euler(x0, y0, xn, h, n):
  p1_euler = [x0]
  p2_euler = [y0]
  
  x = 0
  xfinal = 15

  # questão 1(d) 
  erro = [abs(analitica(x0) - y0) / analitica(x0)]
  
  print('tempo(min)\ttemperatura(°C)')
  while x < xfinal:
    inclinacao = f(x0, y0)
    
    
    yn = y0 + h * inclinacao
      
    x1 = x + h
    x = x1
      
    x0 = x0 + h
    y0 = yn

    print('%.10f\t%.10f\n'% (x0, y0))

    p1_euler.append(x0)
    p2_euler.append(y0)

    # questão 1(d)  
    erro.append(abs(analitica(x0) - yn) / analitica(x0) * 100)
  
  return (p1_euler, p2_euler, erro)
```

```python
p1_euler, p2_euler , erro_euler = euler(0, 6, 15, 0.1, 150)

plt.plot(p1_euler, p2_euler, 'b-', label = "Euler")

plt.xlabel('Tempo (min)')
plt.ylabel('Temperatura (°C)')
plt.title('Euler')
plt.grid(False)

plt.show()
```

Método RK2:
---


```python
def RK2(x0, y0, x, h, n):
  yn = y0

  p1_RK2 = [x0]
  p2_RK2 = [y0]

  # questão 1(d) 
  erro = [abs(analitica(x0) - y0) / analitica(x0)]

  print('tempo(min)\ttemperatura(°C)')
  for i in range(n):
    k1 = h * f(x0, yn)
    k2 = h * f(x0 + 0.5 * h, yn + 0.5 * k1)
        
    yn = yn + (1.0 / 6.0) * (k1 + 2 * k2)
        
    y0 = yn
    x0 = x0 + h

    print('%.10f\t%.10f\n'% (x0, y0))
    
    p1_RK2.append(x0)
    p2_RK2.append(y0)

    # questão 1(d)  
    erro.append(abs(analitica(x0) - yn) / analitica(x0) * 100)
    
  return (p1_RK2, p2_RK2, erro)
```

```python
p1_RK2, p2_RK2, erro_RK2 = RK2(0, 6, 15, 0.1, 150)

plt.plot(p1_RK2, p2_RK2, 'm-', label = "RK2")

plt.xlabel('Tempo (min)')
plt.ylabel('Temperatura (°C)')
plt.title('RK2')
plt.grid(False)

plt.show()
```

Método RK4:
---


```python
def RK4(x0, y0, x, h, n):
  p1_RK4 = [x0]
  p2_RK4 = [y0]

  # questão 1(d) 
  erro = [abs(analitica(x0) - y0) / analitica(x0)]
  
  print('tempo(min)\ttemperatura(°C)') 
  for i in range(n):
    k1 = h * (f(x0, y0))
    k2 = h * (f((x0 + h/ 2), (y0 + k1/ 2)))
    k3 = h * (f((x0 + h/ 2), (y0 + k2/ 2)))
    k4 = h * (f((x0 + h), (y0 + k3)))
       
    k = (k1 + 2*k2 + 2*k3 + k4)/ 6
    yn = y0 + k
        
    y0 = yn
    x0 = x0 + h

    print('%.10f\t%.10f\n'% (x0, y0))

    p1_RK4.append(x0)
    p2_RK4.append(y0)

    # questão 1(d) 
    erro.append(abs(analitica(x0) - yn) / analitica(x0) * 100)
    
  return (p1_RK4, p2_RK4, erro)
```

```python
p1_RK4, p2_RK4, erro_RK4 = RK4(0, 6, 15, 0.1, 150)

plt.plot(p1_RK4, p2_RK4, 'c-', label = "RK4") 

plt.xlabel('Tempo (min)')
plt.ylabel('Temperatura (°C)')
plt.title('RK4')
plt.grid(False)

plt.show()
```

```python
def lotka_volterra_euler(a, c, alpha, gama, h):
  presa = []  # população de presas final
  predador = []   # população de predadores final
  tempo = []   # instante t
 
  contador_tempo = 0
  presa_inicial = 2   # população de presas inicial
  predador_inicial = 5    # população de predadores inicial

  for n in range(int(100/ h)):
    presa_final = presa_inicial + h * (a * presa_inicial - alpha * presa_inicial * predador_inicial)
    predador_final = predador_inicial + h * (-c * predador_inicial + gama * presa_inicial * predador_inicial)
   
    contador_tempo += h
    tempo.append(contador_tempo)

  # não há população negativa
    if(presa_final < 0):
      presa_final = 0

  # não há população negativa
    if(predador_final < 0):
      predador_final = 0

    presa.append(presa_final)
    predador.append(predador_final)

    presa_inicial = presa_final
    predador_inicial = predador_final
    
  print(presa[-1], predador[-1])
  return presa , predador , tempo
```

```python
populacao_presa, populacao_predador, tempo = lotka_volterra_euler(1.0, 0.75, 0.5, 0.5, 0.1)

plt.plot(tempo, populacao_presa, '-m', linestyle ='solid', label = "presa", markersize = 3)
plt.plot(tempo, populacao_predador, '-c', linestyle ='solid', label = "predador", markersize = 3)

plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Lotka-Volterra (Euler)')
plt.legend(loc ="lower right")
plt.grid(False)

plt.show()
```

```python
def lotkaC(a, c, alpha, gama, h, presa_inicial, predador_inicial):
  presa = []  # população de presas final
  predador = []   # população de predadores final
  tempo = []   # instante t
 
  contador_tempo = 0

  for n in range(int(100/ h)):
    presa_final = presa_inicial + h * (a * presa_inicial - alpha * presa_inicial * predador_inicial)
    predador_final = predador_inicial + h * (-c * predador_inicial + gama * presa_inicial * predador_inicial)
   
    contador_tempo += h
    tempo.append(contador_tempo)

  # não há população negativa
    if(presa_final < 0):
      presa_final = 0

  # não há população negativa
    if(predador_final < 0):
      predador_final = 0

    presa.append(presa_final)
    predador.append(predador_final)

    presa_inicial = presa_final
    predador_inicial = predador_final

  return presa , predador , tempo
```

```python
# população inicial de presas maior 
populacao_presa, populacao_predador, t = lotkaC(1.0, 0.75, 0.5, 0.5, 0.001, 6, 2)

plt.plot(t, populacao_presa, '-m', linestyle ='solid', label = "presa", markersize = 3)
plt.plot(t, populacao_predador, '-c', linestyle ='solid', label = "predador", markersize = 3)

plt.xlabel('Tempo')
plt.ylabel('População')
plt.title('Lotka-Volterra (Euler)')
plt.legend(loc ="lower right")
plt.grid(False)

plt.show()
```
